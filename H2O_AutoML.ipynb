```python
import h2o
from h2o.automl import H2OAutoML

h2o.init()
```

```python
data=h2o.import_file("/content/fetal_health classification project.csv")
train, valid, test = data.split_frame(ratios=[0.7, 0.15], seed=42)
target_col = "fetal_health"
features = train.columns
features.remove(target_col)
```

```python
aml = H2OAutoML(max_models=20, seed=42, balance_classes=True, max_runtime_secs=600)
aml.train(x=features, y=target_col, training_frame=train, validation_frame=valid)
```

```python
# View leaderboard
lb = aml.leaderboard
print(lb)
```

```python
# Best model
best_model = aml.leader

# Evaluate the model on the test set
performance = best_model.model_performance(test)
print(performance)
```
